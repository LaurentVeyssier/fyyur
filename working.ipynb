# Dates manipulation

```python
def format_datetime(value:str, format:str='medium')->str:
  """converts a datetime string to a datetime object and returns a formatetd string"""
  # returns a datetime object from a str
  date = dateutil.parser.parse(value)
  if format == 'full':
      format="EEEE MMMM, d, y 'at' h:mma"
  elif format == 'medium':
      format="EE MM, dd, y h:mma"
  # babel takes a datime object and returns a string
  return babel.dates.format_datetime(date, format, locale='en')

app.jinja_env.filters['datetime'] = format_datetime
```

```python
def class_to_dict(cls):
    return {c.key: getattr(cls, c.key) for c in cls.__table__.columns}
```

```python
with app.app_context():
    results=[]
    for city,state in grouped.keys():
        list_venues = grouped[city,state]
        for venue in list_venues:
            venue['num_upcoming_shows' ] = Show.query.filter_by(venue_id=venue['id']).filter(Show.start_time > datetime.now()).count()
        results.append({"city": city, "state": state, "venues": list_venues})
results
```

```python
def show_venue_details(show_id):
    selected_show = Show.query.get(show_id)
    start_time = selected_show.start_time
    artist_id = selected_show.artist_id
    artist = Artist.query.get(artist_id)
    artist_name = artist.name
    artist_image_link = artist.image_link
    return {
        "artist_id": artist_id, 
        "artist_name": artist_name, 
        "artist_image_link": artist_image_link, 
        "start_time": start_time
        }
```
